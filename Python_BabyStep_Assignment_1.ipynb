```python
get_BMI = int(input("Enter the BMI Index:"))
if(get_BMI< 18.5):
    print("Underweight")
elif(18.5 < get_BMI < 24.9):
    print("Normal weight")
elif(25 < get_BMI < 29.9):
    print("Overweight")
elif(30 < get_BMI < 34.9):
    print("Obesity Class I")
elif(35 < get_BMI < 39.9):
    print("Obesity Class II")
elif(get_BMI >= 40):
    print("Obesity Class III")
```
